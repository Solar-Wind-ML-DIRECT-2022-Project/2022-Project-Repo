```python
def uni_selection(df, variable=['ALLSKY_KT']):
    '''for univariate SARIMAX model, creates a new df with solar index'''
    X = df[variable]
    if type(X) != pd.DataFrame:
        raise TypeError('X is pd.Series, not DataFrame. Check parameters.')
    else: pass
    return(X)
```
